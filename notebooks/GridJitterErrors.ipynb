```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from astropy.io import fits
import glob

from mixcoatl.sourcegrid import DistortedGrid
```

```python
e2v_infiles = sorted(glob.glob('/nfs/slac/g/ki/ki19/lsst/snyder18/LSST/Data/BOT/6868D_bf/grid_results/*.fits'))
print(e2v_infiles[0])
itl_infiles = sorted(glob.glob('/nfs/slac/g/ki/ki19/lsst/snyder18/LSST/Data/BOT/6869D_bf/grid_results/*.fits'))
print(itl_infiles[0])
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 12))
axes = axes.flatten()

## Plot E2V results
x0_results = np.zeros(len(e2v_infiles))
y0_results = np.zeros(len(e2v_infiles))
for i, infile in enumerate(e2v_infiles):
    
    grid = DistortedGrid.from_fits(infile)
    y0_results[i] = grid.y0
    x0_results[i] = grid.x0
    
axes[0].plot(x0_results-x0_results[0])
axes[1].plot(y0_results-y0_results[0])

## Plot ITL results
x0_results = np.zeros(len(itl_infiles))
y0_results = np.zeros(len(itl_infiles))
for i, infile in enumerate(itl_infiles):
    
    grid = DistortedGrid.from_fits(infile)
    y0_results[i] = grid.y0
    x0_results[i] = grid.x0
    
axes[2].plot(x0_results-x0_results[0])
axes[3].plot(y0_results-y0_results[0])

axes[0].set_title('E2V Grid Center X', fontsize=18)
axes[0].set_ylim(-10, 10)
axes[0].set_ylabel('Position [pixel]')
axes[0].set_xlabel('Image Number')
axes[0].grid(True, which='major', axis='both')

axes[1].set_title('E2V Grid Center Y', fontsize=18)
axes[1].set_ylim(-10, 10)
axes[1].set_ylabel('Position [pixel]')
axes[1].set_xlabel('Image Number')
axes[1].grid(True, which='major', axis='both')

axes[2].set_title('ITL Grid Center X', fontsize=18)
axes[2].set_ylim(-10, 10)
axes[2].set_ylabel('Position [pixel]')
axes[2].set_xlabel('Image Number')
axes[2].grid(True, which='major', axis='both')

axes[3].set_title('ITL Grid Center Y', fontsize=18)
axes[3].set_ylim(-10, 10)
axes[3].set_ylabel('Position [pixel]')
axes[3].set_xlabel('Image Number')
axes[3].grid(True, which='major', axis='both')

plt.savefig('grid_center_jitter.png')
```
